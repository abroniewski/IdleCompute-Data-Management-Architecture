```python
import pandas as pd
import warnings
import random
from random import randrange
from datetime import timedelta
from datetime import datetime
import socket
import struct
warnings.filterwarnings('ignore')
```

```python
task_num = 5000
user_num = 200
resource_num = 1000
```

```python
user_list = []
for i in range(1,user_num+1):
    user_list.append('U'+str(i))
resource_list = []
for i in range(1,resource_num+1):
    resource_list.append('R'+str(i))
task_list = []
for i in range(1,task_num+1):
    task_list.append('T'+str(i))
```

```python
type_list = ['A1','A2','A3','A4','A5']
type_data = []
for i in range(1,task_num+1):
    type_data.append(type_list[random.randint(0, 4)])
    
data = {'task_id':task_list,'type_id':type_data}
task_type=pd.DataFrame(data,columns=['task_id','type_id'])
task_type.to_csv('task_type.csv',index=False)
task_type
```

```python
def random_date(start, end):
    delta = end - start
    int_delta_hour = ((delta.days * 24 * 60 * 60) + delta.seconds)/3600
    random_hour = randrange(int_delta_hour)
    return start + timedelta(hours=random_hour)
```

```python
year = []
month = []
day = []
task_date_list = []
for i in range(1,task_num+1):
    rand_date = random_date(datetime.strptime('2021-01-01', '%Y-%m-%d'), datetime.strptime('2022-12-31', '%Y-%m-%d'))
    task_date_list.append(rand_date)
    year.append(rand_date.year)
    month.append(rand_date.month)
    day.append(rand_date.day)
    
data = {'task_id':task_list,'year_submitted':year,'month_submitted':month,'day_submitted':day}
task_submission_date=pd.DataFrame(data,columns=['task_id','year_submitted','month_submitted','day_submitted'])
task_submission_date.to_csv('task_submission_date.csv',index=False)
task_submission_date
```

```python
resource_type_list = ['CPU','CPU','GPU']
resource_type_data = []

for i in range(1,resource_num+1):
    resource_type_data.append(resource_type_list[random.randint(0, 2)])
    
data = {'resource_id':resource_list,'resource_type':resource_type_data}
resource_type=pd.DataFrame(data,columns=['resource_id','resource_type'])
resource_type.to_csv('resource_type.csv',index=False)
resource_type
```

```python
ip = []
for i in range(1,resource_num+1):
    ip.append(socket.inet_ntoa(struct.pack('>I', random.randint(1, 0xffffffff))))
    
data = {'resource_id':resource_list,'resource_IP':ip}
resource_IP=pd.DataFrame(data,columns=['resource_id','resource_IP'])
resource_IP.to_csv('resource_IP.csv',index=False)
resource_IP
```

```python
task_rand = []
resource_rand = []
for i in range(0,task_num):
    for j in range(1,random.randint(3, 7)):
        task_rand.append(task_list[i])
        s = random.randint(1, 10)
        if (s==1) or (s==2):
            resource_rand.append('R3') #most frequent
        elif s==3:
            resource_rand.append('R5') #second frequent
        else:
            resource_rand.append(resource_list[random.randint(0, resource_num-1)])

data = {'task_id':task_rand ,'resource_id':resource_rand }
task_resource=pd.DataFrame(data,columns=['task_id','resource_id'])
task_resource.to_csv('task_resource.csv',index=False)
task_resource
```
